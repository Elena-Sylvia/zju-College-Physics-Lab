```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号
plt.rcParams['font.sans-serif'] = ['SimHei']  # 中文字体
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['mathtext.fontset'] = 'cm'  # 使用 Computer Modern 字体渲染数学公式

```

改装电流表

```python
# 数据
experiment_times = np.array([1, 2, 3, 4, 5])
I_modified = np.array([1.00, 2.00, 3.00, 4.00, 5.00])  # 标准电流表读数 /mA
I_standard = np.array([1.06, 2.05, 3.04, 3.95, 5.00])  # 改装安培表读数 /mA
delta_I = np.array([0.06, 0.05, 0.04, -0.05, 0])     # 误差 /mA

# --- 第一张图: DELTA I 关于标准电流表示数变化的折线图 ---
plt.figure(figsize=(8, 6))
plt.plot(I_modified, delta_I, marker='o', linestyle='-', color='blue')
plt.title('$\Delta I$ 关于标准电流表示数的变化', fontsize=16)
plt.xlabel('标准电流表示数 I_标准 (mA)', fontsize=14)
plt.ylabel('误差 $\Delta I$ (mA)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.axhline(0, color='red', linestyle='--', linewidth=0.8, label='零误差线') # 添加零误差线
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

# --- 第二张图: 改装电流以及标准电流关于实验次数的折线图 ---
plt.figure(figsize=(8, 6))
plt.plot(experiment_times, I_modified, marker='o', linestyle='-', color='green', label='标准电流表示数')
plt.plot(experiment_times, I_standard, marker='x', linestyle='--', color='purple', label='改装安培表读数')
plt.title('改装电流与标准电流关系', fontsize=16)
plt.xlabel('实验次数', fontsize=14)
plt.ylabel('电流 (mA)', fontsize=14)
plt.xticks(experiment_times, fontsize=12) # 确保x轴刻度显示为整数实验次数
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

改装欧姆表


```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
Ix = np.array([5.0, 4.6, 4.2, 3.8, 3.4, 3.0, 2.6, 2.2, 1.8, 1.4, 1.0, 0.6, 0.4, 0.2])  # 电流 /mA
Rx = np.array([0, 30.2, 60.2, 96.2, 140.2, 200.1, 260.1, 361.1, 501.1, 701.1, 1101.1, 1801.1, 2643.1, 4443.1]) # 电阻 /Ω

# --- 第一张图: Ix - Rx 曲线 ---
plt.figure(figsize=(9, 6))
plt.plot(Rx, Ix, marker='o', linestyle='-', color='blue')
plt.title('$I_x - R_x$ 曲线', fontsize=16)
plt.xlabel('被测电阻 $R_x$ ($\Omega$)', fontsize=14)
plt.ylabel('电流 $I_x$ (mA)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()

# --- 第二张图: Ix - log(Rx) 曲线 ---

# 过滤掉 Rx = 0 的数据点，因为 log(0) 未定义
# 找到 Rx 不为零的索引
non_zero_indices = Rx != 0
Ix_filtered = Ix[non_zero_indices]
Rx_filtered = Rx[non_zero_indices]

# 计算 log10(Rx)
logRx_filtered = np.log10(Rx_filtered)

plt.figure(figsize=(9, 6))
plt.plot(logRx_filtered, Ix_filtered, marker='x', linestyle='--', color='red')
plt.title('$I_x - \log_{10}(R_x)$ 曲线', fontsize=16)
plt.xlabel('$\log_{10}(R_x)$ ($\Omega$)', fontsize=14)
plt.ylabel('电流 $I_x$ (mA)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
