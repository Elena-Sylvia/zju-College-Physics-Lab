```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 实验数据输入
t_data = np.array([30.0, 35.0, 40.0, 45.0, 50.0, 55.0, 60.0, 65.0])
R_data = np.array([56.99, 58.06, 59.15, 60.23, 61.31, 62.37, 63.46, 64.56])

# 2. 线性拟合 (y = kx + b)
# p[0] 是斜率 k, p[1] 是截距 b
p = np.polyfit(t_data, R_data, 1)
R_fit = np.polyval(p, t_data)

k = p[0]
b = p[1]
alpha_calc = k / b

# 3. 打印计算结果到控制台
print(f"拟合方程: R_t = {k:.4f} * t + {b:.4f}")
print(f"截距 R_0 = {b:.4f} 欧姆")
print(f"斜率 k = {k:.4f}")
print(f"计算出的温度系数 alpha = {alpha_calc:.6f} /℃")

# 4. 绘图
plt.figure(figsize=(8, 6))
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

plt.scatter(t_data, R_data, color='black', label='实验数据点', marker='s')
plt.plot(t_data, R_fit, color='green', linestyle='-', label=f'拟合曲线: $R_t={k:.3f}t + {b:.2f}$')

# 5. 图表装饰
plt.title('铜电阻 $R_t - t$ 温度特性曲线', fontsize=14)
plt.xlabel('温度 t ($^\circ$C)', fontsize=12)
plt.ylabel('电阻 $R_t$ ($\Omega$)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()

# 在图上添加 alpha 值的文本
plt.text(32, 63, f'计算得 $\\alpha = {alpha_calc:.6f} \, ^\circ C^{{-1}}$\n$R^2 \\approx 0.9999$',
         fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```
